```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
close = pd.read_csv('https://raw.githubusercontent.com/HKBU-DavidLo/stock_analysis/master/data/close_prices.csv', index_col=0, parse_dates=True, dayfirst=True)
```

```python
close.head()
```

```python
bdev = close[['BDEV.L']].copy().dropna()
```

```python
bdev.head()
```

```python
bdev.plot();
```

```python
bdev_ret = np.log(bdev.pct_change(1) + 1)
```

```python
bdev_ret.plot();
```

```python
bdev['SMA-10'] = bdev['BDEV.L'].rolling(10).mean() # pandas.core.window.rolling.Rolling
bdev.head(20)
```

```python
bdev['SMA-50'] = bdev['BDEV.L'].rolling(50).mean()
bdev.plot(figsize=(10, 5));
```
