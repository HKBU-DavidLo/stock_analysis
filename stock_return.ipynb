Four property developer stocks listed on London Stcok Exchange as the base dataset, data obtained from yfinance for the period 2018-2022 and stored in csv after removing NA values.

```python
import pandas as pd
```

```python
data_url = 'https://raw.githubusercontent.com/HKBU-DavidLo/stock_analysis/master/data/4stocks.csv'
```

```python
df = pd.read_csv(data_url, index_col="Date", parse_dates=True)
df.head()
```

```python
# select a stock and calculate return as example
bdev = df[df['Name'] == 'BDEV.L']
```

```python
bdev[['Close', 'Adj Close']].plot(figsize=(10, 10))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bdev = df[df['Name'] == 'BDEV.L'].copy()
bdev.head()
```

```python
bdev['Prev Close'] = bdev['Close'].shift(1)
bdev.head()
```

```python
bdev['return'] = bdev['Close'] / bdev['Prev Close'] - 1
bdev.head()
```

```python
bdev['return'].hist(bins=100);
```

```python
bdev['return'].mean(), bdev['return'].std()
```

```python
bdev['log_return'] = np.log(bdev['return'] + 1)
bdev.head()
```

```python
bdev['log_return'].hist(bins=100);
```

```python
bdev['log_return'].mean(), bdev['log_return'].std()
```

Stock price does not reasonably satsify normal distribution because stock price cannot be negative. A better model for stock price is the ***lognormal distribution***. Note: we can say that stock price process is a geometric Brownian motion (or lognormal process) with expected rate of return *mu* and volatility *sigma*. 

# QQ Plot
QQ plots are used to find the type of distribution for a rnadom variable whether it be a Gaussian Distribution, Uniform Distribution, Exponential Distribution or Pareto Distribution.

```python
from scipy.stats import probplot
```

```python
probplot(bdev['return'].dropna(), dist="norm", fit=True, plot=plt);
```

```python
# another method for the QQ plot
import statsmodels.api as sm
```

```python
sm.qqplot(bdev['return'].dropna(), line='s');
```

```python
x_list = np.linspace(
    bdev['log_return'].min(),
    bdev['log_return'].max(),
    100
)
```

```python
from scipy.stats import norm
y_list = norm.pdf(x_list, loc=bdev['log_return'].mean(), scale=bdev['log_return'].std())
```

```python
plt.plot(x_list, y_list);
bdev['log_return'].hist(bins=100, density=True);
```

```python
sm.qqplot(bdev['log_return'].dropna(), line='s');
```

# Covariance and Correlation

```python
data_url = 'https://raw.githubusercontent.com/HKBU-DavidLo/stock_analysis/master/data/close_prices.csv'
close = pd.read_csv(data_url)
```

```python
close.head()
```

```python
close.dropna(axis=0, how='all', inplace=True)
```

```python
symbols = close.columns.tolist()
symbols
```

```python
del symbols[0]
```

```python
symbols
```

```python
for symbol in symbols:
  close[symbol + '_prev'] = close[symbol].shift(1)
  close[symbol + '_ret'] = close[symbol] / close[symbol + '_prev'] - 1
```

```python
rets = close[[symbol + '_ret' for symbol in symbols]].copy()
```

```python
rets.head()
```

```python
import seaborn as sns
sns.pairplot(rets);
```

```python
rets.mean()
```

```python
rets.std()
```

```python
plt.scatter(rets.std(), rets.mean())
plt.xlabel('Vol')
plt.ylabel('Return');
```

```python
rets.corr()
```

```python
rets.cov()
```
